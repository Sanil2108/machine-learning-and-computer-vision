```python
def get_image():
    img = np.array(pyscreenshot.grab())
    left = 38
    width = 800
    top = 600
    height = 220
    return img[top:top+height, left:left+width]
```

```python
cropped = get_image()

fig, ax = plt.subplots(figsize = (10, 10))
ax.imshow(cropped)
```

```python
time_counter_speed = 0
time_counter_turn = 0

time_increment = 0.01

time_speed = 2
time_turn = 0

csv_file = open('data/data.csv', mode='w')
writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

idx = 0

images = []
image_names = []
rows = []

while True:
    if time_counter_turn > time_turn:
        # 0 is straight, 1 is right, 2 is left
        turn = 0
        if keyboard.is_pressed('d'):
            turn = 1
            print('Turning right')
        if keyboard.is_pressed('a'):
            turn = 2
            print('Turning left')
        if keyboard.is_pressed('esc'):
            print('Stopping')
            
#             for i in range(len(images)):
#                 plt.imsave(image_names[i], images[i])
                
#             for i in range(len(rows)):
#                 writer.writerow(rows[i])
            
            csv_file.close()
            
            break;
            
            
        time_counter_turn = 0
        
        image_name = 'data/image_'+(5-len(str(idx)))*'0' + str(idx)+'.png'  
        image = get_image()
        
#         images.append(image)
#         image_names.append(image_name)
    
        writer.writerow([idx, image_name, turn])
        plt.imsave(image_name, image)
        
        idx += 1
    
#     if time_counter_speed > time_speed:
#         keyboard.write('w'*1500)
#         time_counter_speed = 0
    
    time.sleep(time_increment)
    time_counter_speed += time_increment
    time_counter_turn += time_increment
```
