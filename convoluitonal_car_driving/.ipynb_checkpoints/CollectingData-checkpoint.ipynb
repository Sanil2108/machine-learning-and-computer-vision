```python
a = np.array(pyscreenshot.grab())
```

```python
def crop_image(img):
    left = 28
    width = 850
    top = 490
    height = 200
    return img[top:top+height, left:left+width]
```

```python
cropped = crop_image(a)

fig, ax = plt.subplots(figsize = (10, 10))
ax.imshow(cropped)
```

```python
time_counter_speed = 0
time_counter_turn = 0

time_increment = 0.01

time_speed = 3
time_turn = 0.05

while True:
    if time_counter_turn > time_turn:
        if keyboard.is_pressed('d'):
            print('Turning right')
        if keyboard.is_pressed('a'):
            print('Turning left')
        if keyboard.is_pressed('esc'):
            break;
        time_counter_turn = 0
    
    if time_counter_speed > time_speed:
        keyboard.write('w'*1000)
        time_counter_speed = 0
    
    time.sleep(time_increment)
    time_counter_speed += time_increment
    time_counter_turn += time_increment
```
