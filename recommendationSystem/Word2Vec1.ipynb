```python
myData = {}
for obj in data['colors']:
    myData[obj['color']] = np.array([
        int( obj['hex'][1:3], 16 ),
        int( obj['hex'][3:5], 16 ),
        int( obj['hex'][5:7], 16 )
    ])
```

```python
def find_min_distance(idx):
    keys = list(myData.keys())
    vec = myData[keys[idx]]
    lowest = -1
    lowest_dist = -1
    for k, v in myData.items():
        dist = sum((v - vec)**2)
        if (dist < lowest_dist or lowest_dist == -1) and k != keys[idx] :
            lowest_dist = dist
            lowest = k
    return lowest, lowest_dist
```
